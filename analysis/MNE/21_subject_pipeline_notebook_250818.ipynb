## 2. Helper Functions

First, let's create the necessary helper functions for EEG cleaning and feature extraction.

```python
# Automated Channel Cleaning Function
def automated_channel_cleaning(raw, ecg_channel_names=None, z_score_threshold=5.0, 
                              correlation_threshold=0.2, powerline_threshold=15.0,
                              save_report=True, report_path=None):
    """
    Automated EEG channel cleaning: removes ECG channels and detects bad channels.
    """
    
    # Common ECG channel name patterns
    ecg_patterns = ['ECG', 'EKG', 'ecg', 'ekg', 'ECG1', 'ECG2', 'EKG1', 'EKG2']
    
    # Find and remove ECG channels
    ecg_channels_found = []
    if ecg_channel_names is None:
        for ch_name in raw.ch_names:
            if any(pattern in ch_name for pattern in ecg_patterns):
                ecg_channels_found.append(ch_name)
    else:
        ecg_channels_found = ecg_channel_names
    
    # Remove ECG channels
    if ecg_channels_found:
        print(f"Removing ECG channels: {ecg_channels_found}")
        raw = raw.copy().drop_channels(ecg_channels_found)
    else:
        print("No ECG channels found to remove")
        raw = raw.copy()
    
    # Get data for analysis
    data = raw.get_data()
    sfreq = raw.info['sfreq']
    
    bad_channels = []
    
    # 1. Variance-based detection
    channel_vars = np.var(data, axis=1)
    z_scores = np.abs((channel_vars - np.median(channel_vars)) / np.median(channel_vars))
    
    bad_by_variance = []
    for idx, (z_score, ch_name) in enumerate(zip(z_scores, raw.ch_names)):
        if z_score > z_score_threshold:
            bad_by_variance.append(ch_name)
            print(f"  - {ch_name}: High variance (z-score: {z_score:.2f})")
        elif channel_vars[idx] < np.median(channel_vars) * 0.01:
            bad_by_variance.append(ch_name)
            print(f"  - {ch_name}: Nearly flat")
    
    bad_channels.extend(bad_by_variance)
    
    # 2. Correlation-based detection
    downsampled_data = data[:, ::10]
    corr_matrix = np.corrcoef(downsampled_data)
    
    bad_by_correlation = []
    for idx, ch_name in enumerate(raw.ch_names):
        if ch_name in bad_channels:
            continue
        
        correlations = corr_matrix[idx, :]
        correlations[idx] = 0
        mean_corr = np.mean(np.abs(correlations))
        
        if mean_corr < correlation_threshold:
            bad_by_correlation.append(ch_name)
            print(f"  - {ch_name}: Low correlation (mean: {mean_corr:.2f})")
    
    bad_channels.extend(bad_by_correlation)
    
    # Remove duplicates
    bad_channels = list(set(bad_channels))
    
    # Mark bad channels
    raw.info['bads'] = bad_channels
    
    print(f"\nTotal bad channels detected: {len(bad_channels)}")
    
    # Simple QC stats
    qc_stats = {
        'bad_channels': bad_channels,
        'bad_by_variance': bad_by_variance,
        'bad_by_correlation': bad_by_correlation,
        'ecg_channels_removed': ecg_channels_found
    }
    
    # Optional: Save a simple plot
    if save_report and report_path:
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.bar(range(len(channel_vars)), channel_vars)
        ax.set_xlabel('Channel')
        ax.set_ylabel('Variance')
        ax.set_title('Channel Variance Distribution')
        for i, ch in enumerate(bad_channels):
            if ch in raw.ch_names:
                idx = raw.ch_names.index(ch)
                ax.bar(idx, channel_vars[idx], color='red')
        plt.tight_layout()
        plt.savefig(report_path)
        plt.close()
    
    return raw, bad_channels, qc_stats
```

```python
# Lempel-Ziv Complexity Function
def compute_lz_features_matlab_style(eeg_data, fs, window_length=1.0, overlap=0.5, 
                                   complexity_type='exhaustive', use_fast=True):
    """
    Compute Lempel-Ziv complexity features from EEG data.
    """
    
    n_channels, n_samples = eeg_data.shape
    window_samples = int(window_length * fs)
    step_samples = int(window_samples * (1 - overlap))
    n_windows = (n_samples - window_samples) // step_samples + 1
    
    print(f"Computing LZ complexity ({complexity_type})...")
    
    # Pre-allocate arrays
    lz_complexity = np.zeros((n_windows, n_channels))
    lz_times = np.zeros(n_windows)
    
    # Pre-compute channel medians for binarization
    channel_medians = np.median(eeg_data, axis=1)
    
    # Process windows
    for win_idx in tqdm(range(n_windows), desc=f"LZ {complexity_type}"):
        start = win_idx * step_samples
        end = start + window_samples
        
        if end > n_samples:
            break
            
        lz_times[win_idx] = start / fs + window_length / 2
        
        # Compute LZ for each channel
        for ch in range(n_channels):
            # Extract and binarize window
            window_data = eeg_data[ch, start:end]
            binary_seq = (window_data > channel_medians[ch]).astype(int)
            
            # Simplified LZ complexity calculation
            binary_str = ''.join(binary_seq.astype(str))
            
            # Count unique substrings (simplified approach)
            substrings = set()
            max_len = min(10, len(binary_str) // 4)
            
            for length in range(1, max_len + 1):
                for i in range(len(binary_str) - length + 1):
                    substrings.add(binary_str[i:i+length])
            
            c = len(substrings)
            if len(binary_str) > 1:
                c_norm = c / (len(binary_str) / np.log2(len(binary_str)))
            else:
                c_norm = c
            
            lz_complexity[win_idx, ch] = c_norm
    
    # Trim to actual number of windows
    actual_windows = win_idx
    
    features = {
        'lz_complexity_mean': np.mean(lz_complexity[:actual_windows], axis=1),
        'lz_complexity_std': np.std(lz_complexity[:actual_windows], axis=1),
        'lz_complexity_max': np.max(lz_complexity[:actual_windows], axis=1),
        'lz_complexity_min': np.min(lz_complexity[:actual_windows], axis=1),
        'lz_times': lz_times[:actual_windows]
    }
    
    # Add spatial features if enough channels
    if n_channels >= 20:
        features['lz_complexity_frontal'] = np.mean(lz_complexity[:actual_windows, :10], axis=1)
        features['lz_complexity_posterior'] = np.mean(lz_complexity[:actual_windows, -10:], axis=1)
        features['lz_frontal_posterior_diff'] = (
            features['lz_complexity_frontal'] - features['lz_complexity_posterior']
        )
    
    return features
```

## 3. Main Pipeline Class

This is the core pipeline class that handles all processing steps.

```python
class MultiSubjectPDAPipeline:
    """
    Complete pipeline for multi-subject PDA prediction from EEG data.
    """
    
    def __init__(self, base_dir='./data', output_dir='./results'):
        """
        Initialize the pipeline.
        """
        self.base_dir = base_dir
        self.output_dir = output_dir
        self.subjects_data = {}
        self.combined_features = None
        self.combined_targets = None
        self.subject_ids = None
        self.trained_model = None
        self.scaler = None
        
        # Create output directory
        os.makedirs(output_dir, exist_ok=True)
        
        # Pipeline parameters
        self.params = {
            'eeg': {
                'z_score_threshold': 5.0,
                'correlation_threshold': 0.2,
                'powerline_threshold': 15.0
            },
            'pda': {
                'hrf_delay': 5.0,
                'target_fs': 1.0
            },
            'stft': {
                'window_sec': 1.0,
                'overlap': 0.5
            },
            'lz': {
                'window_length': 2.0,
                'overlap': 0.5,
                'complexity_type': 'exhaustive',
                'use_fast': True
            },
            'features': {
                'n_top_features': 20,
                'use_lagged_features': True,
                'lag_samples': [1, 2, 3, 4, 5, 6]
            },
            'model': {
                'test_size': 0.2,
                'cv_folds': 5,
                'random_state': 42
            }
        }
    
    def process_single_subject(self, subject_id, eeg_file, pda_file):
        """
        Process a single subject's data.
        """
        print(f"\n{'='*70}")
        print(f"Processing Subject: {subject_id}")
        print(f"{'='*70}")
        
        subject_output_dir = os.path.join(self.output_dir, subject_id)
        os.makedirs(subject_output_dir, exist_ok=True)
        
        # 1. Load and clean EEG data
        print("\n1. Loading and cleaning EEG data...")
        raw_processed, bad_channels, qc_stats = self._clean_eeg_data(
            eeg_file, subject_output_dir
        )
        
        # 2. Load PDA data
        print("\n2. Loading PDA data...")
        pda_df = pd.read_csv(pda_file)
        
        # Calculate PDA (CEN - DMN)
        if 'cen' in pda_df.columns and 'dmn' in pda_df.columns:
            pda_signal = pda_df['cen'].values - pda_df['dmn'].values
        else:
            raise ValueError(f"PDA file must contain 'cen' and 'dmn' columns")
        
        # 3. Align PDA with EEG
        print("\n3. Aligning PDA with EEG...")
        pda_aligned_z, pda_time_aligned = self._align_pda_to_eeg(
            pda_signal, raw_processed
        )
        
        # 4. Extract features
        print("\n4. Extracting features...")
        
        # 4a. STFT features
        stft_features = self._extract_stft_features(
            raw_processed, pda_aligned_z, pda_time_aligned
        )
        
        # 4b. LZ complexity features
        lz_features = self._extract_lz_features(
            raw_processed, pda_aligned_z, pda_time_aligned
        )
        
        # 4c. Advanced features
        advanced_features = self._extract_advanced_features(
            raw_processed, stft_features, pda_aligned_z, pda_time_aligned
        )
        
        # 5. Combine all features
        print("\n5. Combining features...")
        all_features = self._combine_features(
            stft_features, lz_features, advanced_features
        )
        
        # 6. Save subject data
        subject_data = {
            'subject_id': subject_id,
            'features': all_features['feature_matrix'],
            'feature_names': all_features['feature_names'],
            'target': pda_aligned_z,
            'time_points': pda_time_aligned,
            'eeg_info': {
                'n_channels': len(raw_processed.ch_names),
                'bad_channels': bad_channels,
                'duration': raw_processed.times[-1]
            }
        }
        
        # Save to file
        np.savez(
            os.path.join(subject_output_dir, f'{subject_id}_processed.npz'),
            **subject_data
        )
        
        print(f"\nSubject {subject_id} processing complete!")
        print(f"Features shape: {subject_data['features'].shape}")
        print(f"Target shape: {subject_data['target'].shape}")
        
        return subject_data
    
    def _clean_eeg_data(self, eeg_file, output_dir):
        """Clean EEG data using automated bad channel detection."""
        # Load raw data
        raw = mne.io.read_raw_edf(eeg_file, preload=True)
        
        # Run automated cleaning
        raw_cleaned, bad_channels, qc_stats = automated_channel_cleaning(
            raw,
            z_score_threshold=self.params['eeg']['z_score_threshold'],
            correlation_threshold=self.params['eeg']['correlation_threshold'],
            powerline_threshold=self.params['eeg']['powerline_threshold'],
            save_report=True,
            report_path=os.path.join(output_dir, 'qc_report.png')
        )
        
        # Remove bad channels
        good_channels = [ch for ch in raw_cleaned.ch_names if ch not in bad_channels]
        raw_processed = raw_cleaned.copy().pick_channels(good_channels)
        
        return raw_processed, bad_channels, qc_stats
    
    def _align_pda_to_eeg(self, pda_signal, raw_data):
        """Align PDA signal to EEG timeline with HRF delay."""
        # Original PDA timing
        pda_fs = 1 / 1.2  # ~0.833 Hz
        pda_time_orig = np.arange(len(pda_signal)) * 1.2
        
        # Apply HRF shift
        hrf_delay = self.params['pda']['hrf_delay']
        pda_time_shifted = pda_time_orig - hrf_delay
        
        # Target time points
        eeg_duration = raw_data.times[-1]
        target_fs = self.params['pda']['target_fs']
        target_time = np.arange(0, min(eeg_duration, pda_time_shifted[-1]), 1/target_fs)
        
        # Only use valid PDA samples
        valid_idx = pda_time_shifted >= 0
        
        # Interpolate
        interp_func = interp1d(
            pda_time_shifted[valid_idx], 
            pda_signal[valid_idx], 
            kind='cubic',
            bounds_error=False,
            fill_value='extrapolate'
        )
        pda_resampled = interp_func(target_time)
        
        # Z-score normalization
        pda_resampled_z = (pda_resampled - np.mean(pda_resampled)) / np.std(pda_resampled)
        
        return pda_resampled_z, target_time
    
    def _extract_stft_features(self, raw_processed, pda_aligned_z, pda_time_aligned):
        """Extract STFT-based band power features."""
        fs = raw_processed.info['sfreq']
        
        # STFT parameters
        window_sec = self.params['stft']['window_sec']
        nperseg = int(window_sec * fs)
        noverlap = int(self.params['stft']['overlap'] * fs)
        
        # Get EEG data
        picks_eeg = mne.pick_types(raw_processed.info, eeg=True)
        eeg_data = raw_processed.get_data(picks=picks_eeg)
        
        # Compute STFT
        f, t_stft, Zxx = stft(eeg_data, fs=fs, nperseg=nperseg, 
                             noverlap=noverlap, axis=1)
        
        # Calculate power
        power = np.abs(Zxx) ** 2
        power = power.transpose(2, 0, 1)  # (n_windows, n_channels, n_freqs)
        
        # Extract band powers
        bands = {
            'theta': (4, 8),
            'alpha': (8, 13),
            'beta': (13, 30),
            'low_gamma': (30, 50)
        }
        
        band_powers = {}
        for band_name, (low, high) in bands.items():
            band_mask = (f >= low) & (f <= high)
            band_powers[band_name] = np.mean(power[:, :, band_mask], axis=2)
        
        # Add ratio features
        eps = 1e-10
        band_powers['theta_alpha_ratio'] = band_powers['theta'] / (band_powers['alpha'] + eps)
        band_powers['beta_alpha_ratio'] = band_powers['beta'] / (band_powers['alpha'] + eps)
        
        return {
            'band_powers': band_powers,
            'stft_times': t_stft,
            'power': power,
            'freqs': f
        }
    
    def _extract_lz_features(self, raw_processed, pda_aligned_z, pda_time_aligned):
        """Extract Lempel-Ziv complexity features."""
        picks_eeg = mne.pick_types(raw_processed.info, eeg=True)
        eeg_data = raw_processed.get_data(picks=picks_eeg)
        fs = raw_processed.info['sfreq']
        
        # Compute LZ complexity
        features = compute_lz_features_matlab_style(
            eeg_data, fs,
            window_length=self.params['lz']['window_length'],
            overlap=self.params['lz']['overlap'],
            complexity_type=self.params['lz']['complexity_type'],
            use_fast=self.params['lz']['use_fast']
        )
        
        return features
    
    def _extract_advanced_features(self, raw_processed, stft_features, 
                                  pda_aligned_z, pda_time_aligned):
        """Extract advanced features including spatial and connectivity."""
        fs = raw_processed.info['sfreq']
        picks_eeg = mne.pick_types(raw_processed.info, eeg=True)
        channel_names = [raw_processed.ch_names[i] for i in picks_eeg]
        
        band_powers = stft_features['band_powers']
        stft_times = stft_features['stft_times']
        
        features = {}
        
        # Spatial features
        frontal_channels = []
        parietal_channels = []
        
        for i, ch in enumerate(channel_names):
            ch_upper = ch.upper()
            if any(x in ch_upper for x in ['FP1', 'FP2', 'F3', 'F4', 'F7', 'F8', 'FZ']):
                frontal_channels.append(i)
            if any(x in ch_upper for x in ['P3', 'P4', 'PZ', 'P7', 'P8']):
                parietal_channels.append(i)
        
        # Network indices
        if frontal_channels and parietal_channels:
            frontal_alpha = np.mean(band_powers['alpha'][:, frontal_channels], axis=1)
            parietal_alpha = np.mean(band_powers['alpha'][:, parietal_channels], axis=1)
            features['frontal_parietal_alpha_gradient'] = parietal_alpha - frontal_alpha
            
            dmn_index = (np.mean(band_powers['alpha'][:, parietal_channels], axis=1) - 
                        np.mean(band_powers['beta'][:, frontal_channels], axis=1))
            cen_index = (np.mean(band_powers['beta'][:, frontal_channels], axis=1) - 
                        np.mean(band_powers['alpha'][:, parietal_channels], axis=1))
        else:
            dmn_index = np.mean(band_powers['alpha'], axis=1) - np.mean(band_powers['beta'], axis=1)
            cen_index = np.mean(band_powers['beta'], axis=1) - np.mean(band_powers['alpha'], axis=1)
        
        features['dmn_index'] = dmn_index
        features['cen_index'] = cen_index
        features['network_competition'] = cen_index - dmn_index
        
        return {
            'features': features,
            'times': stft_times
        }
    
    def _combine_features(self, stft_features, lz_features, advanced_features):
        """Combine all features and align with PDA."""
        all_aligned_features = {}
        
        # Add STFT band powers
        for band_name, band_data in stft_features['band_powers'].items():
            # Mean across channels
            feat_mean = np.mean(band_data, axis=1)
            all_aligned_features[f'{band_name}_mean'] = feat_mean
            
            # Std across channels
            feat_std = np.std(band_data, axis=1)
            all_aligned_features[f'{band_name}_std'] = feat_std
        
        # Add LZ features
        for feat_name, feat_data in lz_features.items():
            if feat_name != 'lz_times' and isinstance(feat_data, np.ndarray):
                all_aligned_features[feat_name] = feat_data
        
        # Add advanced features
        for feat_name, feat_data in advanced_features['features'].items():
            if isinstance(feat_data, np.ndarray):
                all_aligned_features[feat_name] = feat_data
        
        # Get common time array (use STFT times as reference)
        ref_times = stft_features['stft_times']
        
        # Create feature matrix
        feature_list = []
        feature_names = []
        
        for feat_name, feat_data in all_aligned_features.items():
            if feat_data.ndim == 1 and len(feat_data) == len(ref_times):
                feature_list.append(feat_data.reshape(-1, 1))
                feature_names.append(feat_name)
        
        feature_matrix = np.hstack(feature_list)
        
        # Add lagged features if requested
        if self.params['features']['use_lagged_features']:
            lag_samples = self.params['features']['lag_samples']
            key_features = ['network_competition', 'cen_index', 'dmn_index', 
                          'beta_alpha_ratio_mean', 'lz_complexity_mean']
            
            for lag in lag_samples:
                for feat in key_features:
                    if feat in feature_names:
                        feat_idx = feature_names.index(feat)
                        lagged_feat = np.roll(feature_matrix[:, feat_idx], lag)
                        lagged_feat[:lag] = lagged_feat[lag]
                        
                        feature_list.append(lagged_feat.reshape(-1, 1))
                        feature_names.append(f'{feat}_lag{lag}s')
            
            feature_matrix = np.hstack(feature_list)
        
        return {
            'feature_matrix': feature_matrix,
            'feature_names': feature_names
        }
    
    def process_all_subjects(self, subject_files):
        """
        Process all subjects in the dataset.
        """
        print(f"\n{'='*70}")
        print(f"PROCESSING {len(subject_files)} SUBJECTS")
        print(f"{'='*70}")
        
        for subject_id, eeg_file, pda_file in subject_files:
            try:
                subject_data = self.process_single_subject(
                    subject_id, eeg_file, pda_file
                )
                self.subjects_data[subject_id] = subject_data
            except Exception as e:
                print(f"\nError processing subject {subject_id}: {e}")
                continue
        
        print(f"\n\nSuccessfully processed {len(self.subjects_data)} subjects")
    
    def combine_subjects_data(self):
        """Combine data from all subjects for training."""
        print(f"\n{'='*70}")
        print("COMBINING MULTI-SUBJECT DATA")
        print(f"{'='*70}")
        
        all_features = []
        all_targets = []
        all_subject_ids = []
        
        # Get common feature names (use first subject as reference)
        ref_subject = list(self.subjects_data.values())[0]
        common_features = ref_subject['feature_names']
        
        for subject_id, data in self.subjects_data.items():
            # Ensure feature consistency
            if data['feature_names'] == common_features:
                all_features.append(data['features'])
                all_targets.append(data['target'])
                all_subject_ids.extend([subject_id] * len(data['target']))
            else:
                print(f"Warning: Feature mismatch for subject {subject_id}")
        
        # Combine arrays
        self.combined_features = np.vstack(all_features)
        self.combined_targets = np.hstack(all_targets)
        self.subject_ids = np.array(all_subject_ids)
        self.feature_names = common_features
        
        print(f"\nCombined data shape:")
        print(f"  Features: {self.combined_features.shape}")
        print(f"  Targets: {self.combined_targets.shape}")
        print(f"  Subjects: {len(np.unique(self.subject_ids))}")
    
    def train_models(self):
        """Train and evaluate multiple models with cross-validation."""
        print(f"\n{'='*70}")
        print("TRAINING MULTI-SUBJECT MODELS")
        print(f"{'='*70}")
        
        # Feature selection
        top_features, correlations = self._select_top_features()
        
        # Create feature matrix with selected features
        feature_indices = [self.feature_names.index(f) for f in top_features]
        X = self.combined_features[:, feature_indices]
        y = self.combined_targets
        groups = self.subject_ids
        
        # Standardize features
        self.scaler = StandardScaler()
        X_scaled = self.scaler.fit_transform(X)
        
        # Models to evaluate
        models = {
            'Ridge': Ridge(alpha=1.0),
            'ElasticNet': ElasticNet(alpha=0.1, l1_ratio=0.5, max_iter=2000),
            'RandomForest': RandomForestRegressor(
                n_estimators=100, max_depth=5, random_state=42, n_jobs=-1
            ),
            'XGBoost': xgb.XGBRegressor(
                n_estimators=100, max_depth=3, learning_rate=0.1,
                random_state=42, n_jobs=-1
            )
        }
        
        # Group K-Fold for subject-wise cross-validation
        gkf = GroupKFold(n_splits=self.params['model']['cv_folds'])
        
        results = {}
        
        print("\nCross-Validation Results:")
        print("-" * 50)
        
        for model_name, model in models.items():
            # Cross-validation with groups
            cv_scores = []
            
            for train_idx, val_idx in gkf.split(X_scaled, y, groups):
                X_train, X_val = X_scaled[train_idx], X_scaled[val_idx]
                y_train, y_val = y[train_idx], y[val_idx]
                
                model.fit(X_train, y_train)
                y_pred = model.predict(X_val)
                
                corr, _ = pearsonr(y_val, y_pred)
                cv_scores.append(corr)
            
            cv_scores = np.array(cv_scores)
            
            # Fit on full data
            model.fit(X_scaled, y)
            y_pred_full = model.predict(X_scaled)
            
            results[model_name] = {
                'model': model,
                'cv_scores': cv_scores,
                'cv_mean': cv_scores.mean(),
                'cv_std': cv_scores.std(),
                'full_correlation': pearsonr(y, y_pred_full)[0],
                'full_r2': r2_score(y, y_pred_full)
            }
            
            print(f"\n{model_name}:")
            print(f"  CV correlations: {cv_scores.round(3)}")
            print(f"  Mean CV r: {cv_scores.mean():.3f} (±{cv_scores.std():.3f})")
            print(f"  Full data r: {results[model_name]['full_correlation']:.3f}")
        
        # Select best model
        best_model_name = max(results.keys(), 
                            key=lambda x: results[x]['cv_mean'])
        
        print(f"\n{'='*50}")
        print(f"BEST MODEL: {best_model_name}")
        print(f"Mean CV correlation: {results[best_model_name]['cv_mean']:.3f}")
        print(f"{'='*50}")
        
        # Store results
        self.model_results = results
        self.best_model = results[best_model_name]['model']
        self.selected_features = top_features
        
        # Create visualization
        self._plot_model_results(results, X_scaled, y, groups)
        
        return results
    
    def _select_top_features(self):
        """Select top features based on correlation with target."""
        correlations = {}
        
        for i, feat_name in enumerate(self.feature_names):
            feat_data = self.combined_features[:, i]
            if np.std(feat_data) > 1e-10:
                corr, _ = pearsonr(feat_data, self.combined_targets)
                correlations[feat_name] = abs(corr)
        
        # Sort by correlation
        sorted_features = sorted(correlations.items(), 
                               key=lambda x: x[1], reverse=True)
        
        # Get top N features
        n_features = self.params['features']['n_top_features']
        top_features = [f[0] for f in sorted_features[:n_features]]
        
        print(f"\nTop {n_features} features:")
        for i, (feat, corr) in enumerate(sorted_features[:n_features]):
            print(f"{i+1:2d}. {feat:<40} |r|={corr:.3f}")
        
        return top_features, correlations
    
    def _plot_model_results(self, results, X_scaled, y, groups):
        """Create visualization of model results."""
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Multi-Subject Model Results', fontsize=16)
        
        # 1. Model comparison
        ax = axes[0, 0]
        model_names = list(results.keys())
        cv_means = [results[m]['cv_mean'] for m in model_names]
        cv_stds = [results[m]['cv_std'] for m in model_names]
        
        bars = ax.bar(model_names, cv_means, yerr=cv_stds, capsize=5)
        ax.set_ylabel('Mean CV Correlation')
        ax.set_title('Model Performance Comparison')
        ax.grid(True, alpha=0.3, axis='y')
        
        # Highlight best model
        best_idx = np.argmax(cv_means)
        bars[best_idx].set_color('darkgreen')
        
        # 2. Subject-wise performance
        ax = axes[0, 1]
        best_model = self.best_model
        
        subject_correlations = {}
        unique_subjects = np.unique(groups)
        
        for subject in unique_subjects:
            subject_mask = groups == subject
            X_subj = X_scaled[subject_mask]
            y_subj = y[subject_mask]
            
            y_pred = best_model.predict(X_subj)
            corr, _ = pearsonr(y_subj, y_pred)
            subject_correlations[subject] = corr
        
        subjects = list(subject_correlations.keys())
        correlations = list(subject_correlations.values())
        
        ax.bar(range(len(subjects)), correlations)
        ax.set_xlabel('Subject')
        ax.set_ylabel('Correlation')
        ax.set_title('Per-Subject Performance (Best Model)')
        ax.set_xticks(range(len(subjects)))
        ax.set_xticklabels(subjects, rotation=45)
        ax.grid(True, alpha=0.3, axis='y')
        
        # 3. Feature importance (for tree-based models)
        ax = axes[1, 0]
        if hasattr(self.best_model, 'feature_importances_'):
            importances = self.best_model.feature_importances_
            indices = np.argsort(importances)[::-1][:10]
            
            top_features = [self.selected_features[i] for i in indices]
            top_importances = importances[indices]
            
            y_pos = np.arange(len(top_features))
            ax.barh(y_pos, top_importances)
            ax.set_yticks(y_pos)
            ax.set_yticklabels(top_features)
            ax.set_xlabel('Importance')
            ax.set_title('Top 10 Feature Importances')
            ax.grid(True, alpha=0.3, axis='x')
        else:
            ax.text(0.5, 0.5, 'Feature importance\nnot available\nfor this model',
                   ha='center', va='center', transform=ax.transAxes)
        
        # 4. Predictions scatter plot
        ax = axes[1, 1]
        y_pred = self.best_model.predict(X_scaled)
        
        # Downsample for clarity
        downsample = 10
        ax.scatter(y[::downsample], y_pred[::downsample], 
                  alpha=0.5, s=20)
        ax.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
        ax.set_xlabel('Actual PDA')
        ax.set_ylabel('Predicted PDA')
        ax.set_title(f'Predictions (r={pearsonr(y, y_pred)[0]:.3f})')
        ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(os.path.join(self.output_dir, 'model_results.png'), 
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    def test_on_new_subject(self, subject_id, eeg_file, pda_file=None):
        """
        Test the trained model on a new subject.
        """
        print(f"\n{'='*70}")
        print(f"TESTING ON NEW SUBJECT: {subject_id}")
        print(f"{'='*70}")
        
        # Process the subject
        test_data = self.process_single_subject(subject_id, eeg_file, pda_file)
        
        # Extract selected features
        feature_indices = [test_data['feature_names'].index(f) 
                          for f in self.selected_features]
        X_test = test_data['features'][:, feature_indices]
        
        # Scale features
        X_test_scaled = self.scaler.transform(X_test)
        
        # Predict
        predictions = self.best_model.predict(X_test_scaled)
        
        results = {
            'subject_id': subject_id,
            'predictions': predictions,
            'time_points': test_data['time_points']
        }
        
        # If true PDA provided, evaluate
        if pda_file is not None:
            y_true = test_data['target']
            correlation, _ = pearsonr(y_true, predictions)
            r2 = r2_score(y_true, predictions)
            rmse = np.sqrt(mean_squared_error(y_true, predictions))
            
            results.update({
                'true_pda': y_true,
                'correlation': correlation,
                'r2': r2,
                'rmse': rmse
            })
            
            print(f"\nTest Results:")
            print(f"  Correlation: {correlation:.3f}")
            print(f"  R²: {r2:.3f}")
            print(f"  RMSE: {rmse:.3f}")
            
            # Plot results
            self._plot_test_results(results)
        
        return results
    
    def _plot_test_results(self, results):
        """Plot test results for a single subject."""
        fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
        
        time_points = results['time_points']
        predictions = results['predictions']
        true_pda = results['true_pda']
        
        # Time series
        ax = axes[0]
        ax.plot(time_points, true_pda, 'k-', label='True PDA', alpha=0.8)
        ax.plot(time_points, predictions, 'r--', label='Predicted PDA', alpha=0.8)
        ax.set_ylabel('PDA (z-score)')
        ax.set_title(f"Test Subject: {results['subject_id']} (r={results['correlation']:.3f})")
        ax.legend()
        ax.grid(True, alpha=0.3)
        
        # Residuals
        ax = axes[1]
        residuals = true_pda - predictions
        ax.plot(time_points, residuals, 'g-', alpha=0.7)
        ax.axhline(y=0, color='k', linestyle='--', alpha=0.5)
        ax.set_xlabel('Time (s)')
        ax.set_ylabel('Residuals')
        ax.set_title('Prediction Errors')
        ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(os.path.join(self.output_dir, f"test_{results['subject_id']}.png"),
                   dpi=300, bbox_inches='tight')
        plt.show()
    
    def save_model(self, filename='multi_subject_pda_model.pkl'):
        """Save the trained model and necessary components."""
        model_data = {
            'model': self.best_model,
            'scaler': self.scaler,
            'selected_features': self.selected_features,
            'model_results': self.model_results,
            'params': self.params,
            'n_subjects_trained': len(np.unique(self.subject_ids))
        }
        
        filepath = os.path.join(self.output_dir, filename)
        joblib.dump(model_data, filepath)
        print(f"\nModel saved to: {filepath}")
    
    def load_model(self, filepath):
        """Load a previously trained model."""
        model_data = joblib.load(filepath)
        self.best_model = model_data['model']
        self.scaler = model_data['scaler']
        self.selected_features = model_data['selected_features']
        self.model_results = model_data['model_results']
        self.params = model_data['params']
        print(f"Model loaded from: {filepath}")
        print(f"Trained on {model_data['n_subjects_trained']} subjects")
```

```python
# Define the diagnostic combine function
def combine_subjects_data_diagnostic(self):
    """Diagnostic version: Combine data from all subjects with detailed error checking."""
    print(f"\n{'='*70}")
    print("DIAGNOSTIC: COMBINING MULTI-SUBJECT DATA")
    print(f"{'='*70}")
    
    # First, let's examine the data structure
    print("\n1. EXAMINING DATA STRUCTURE:")
    print("-" * 40)
    
    if not hasattr(self, 'subjects_data') or not self.subjects_data:
        print("ERROR: No subjects_data found!")
        return
    
    print(f"Number of subjects: {len(self.subjects_data)}")
    
    all_features = []
    all_targets = []
    all_subject_ids = []
    
    # Get reference subject for feature names
    ref_subject_id = list(self.subjects_data.keys())[0]
    ref_data = self.subjects_data[ref_subject_id]
    
    if 'feature_names' not in ref_data:
        print("ERROR: No feature_names in reference subject!")
        return
    else:
        common_features = ref_data['feature_names']
    
    # Process each subject
    successful_subjects = 0
    for subject_id, data in self.subjects_data.items():
        try:
            # Get features and target
            if 'features' not in data or 'target' not in data:
                print(f"WARNING: Skipping {subject_id} - missing features or target")
                continue
            
            features = data['features']
            target = data['target']
            
            # Convert to numpy arrays if needed
            if not isinstance(features, np.ndarray):
                features = np.array(features)
            if not isinstance(target, np.ndarray):
                target = np.array(target)
            
            # Handle different dimensionalities
            if len(features.shape) == 1:
                features = features.reshape(-1, 1)
            
            # Get lengths
            n_feature_samples = features.shape[0]
            n_target_samples = len(target) if hasattr(target, '__len__') else 1
            
            # Find minimum length
            min_length = min(n_feature_samples, n_target_samples)
            
            if min_length == 0:
                print(f"  WARNING: Zero length data for {subject_id}, skipping!")
                continue
            
            # Trim to same length
            features_trimmed = features[:min_length]
            target_trimmed = target[:min_length] if hasattr(target, '__getitem__') else np.array([target])
            
            # Add to lists
            all_features.append(features_trimmed)
            all_targets.append(target_trimmed)
            all_subject_ids.extend([subject_id] * min_length)
            
            successful_subjects += 1
            print(f"  Added {subject_id}: {min_length} samples")
            
        except Exception as e:
            print(f"ERROR processing {subject_id}: {str(e)}")
            import traceback
            traceback.print_exc()
    
    # Combine arrays
    if successful_subjects > 0:
        try:
            self.combined_features = np.vstack(all_features)
            self.combined_targets = np.hstack(all_targets)
            self.subject_ids = np.array(all_subject_ids)
            self.feature_names = common_features
            
            print(f"\n2. COMBINATION SUCCESSFUL:")
            print("-" * 40)
            print(f"  Combined features: {self.combined_features.shape}")
            print(f"  Combined targets: {self.combined_targets.shape}")
            print(f"  Subject IDs: {len(self.subject_ids)}")
            print(f"  Unique subjects: {len(np.unique(self.subject_ids))}")
            
        except Exception as e:
            print(f"\nERROR during combination: {str(e)}")
            import traceback
            traceback.print_exc()
    else:
        print("\nERROR: No subjects were successfully processed!")

# Apply the fix
MultiSubjectPDAPipeline.combine_subjects_data = combine_subjects_data_diagnostic

# Now analyze variability across all subjects
print("="*70)
print("ANALYZING VARIABILITY ACROSS ALL AVAILABLE SUBJECTS")
print("="*70)

# Get fresh subject list
all_subject_runs = prepare_subject_list()
print(f"\nTotal available subject-runs: {len(all_subject_runs)}")

# Create new pipeline for all subjects
pipeline_all = MultiSubjectPDAPipeline(
    base_dir='../',
    output_dir='./results_all_subjects'
)

# Process all subjects (this will take a while)
print("\nProcessing all subjects...")
print("(This may take 20-30 minutes for 21 subjects)")

# Process in batches to see progress
batch_size = 5
for i in range(0, len(all_subject_runs), batch_size):
    batch = all_subject_runs[i:i+batch_size]
    print(f"\nProcessing batch {i//batch_size + 1}/{(len(all_subject_runs)-1)//batch_size + 1}")
    pipeline_all.process_all_subjects(batch)

# Combine all data
print("\nCombining all subjects...")
pipeline_all.combine_subjects_data()

# Now do the variability analysis
if hasattr(pipeline_all, 'combined_features') and pipeline_all.combined_features is not None:
    print("\n" + "="*70)
    print("SUBJECT VARIABILITY SUMMARY")
    print("="*70)
    
    unique_subjects = np.unique(pipeline_all.subject_ids)
    print(f"\nSuccessfully processed {len(unique_subjects)} subjects")
    
    # Quick summary of each subject
    print("\nSubject summary:")
    print(f"{'Subject':<25} {'N samples':>10} {'PDA mean':>10} {'PDA std':>10}")
    print("-" * 60)
    
    for subject in unique_subjects:
        subject_mask = pipeline_all.subject_ids == subject
        pda_data = pipeline_all.combined_targets[subject_mask]
        print(f"{subject:<25} {np.sum(subject_mask):>10d} {np.mean(pda_data):>10.3f} {np.std(pda_data):>10.3f}")
    
    # Save the processed data for later use
    np.savez('all_subjects_processed.npz',
             features=pipeline_all.combined_features,
             targets=pipeline_all.combined_targets,
             subject_ids=pipeline_all.subject_ids,
             feature_names=pipeline_all.feature_names)
    print("\nSaved all processed data to 'all_subjects_processed.npz'")
else:
    print("ERROR: No data was successfully combined")
```

```python
# Load the saved data for analysis
data = np.load('all_subjects_processed.npz', allow_pickle=True)
features = data['features']
targets = data['targets']
subject_ids = data['subject_ids']
feature_names = data['feature_names']

print("="*70)
print("BETWEEN-SUBJECT VARIABILITY ANALYSIS")
print("="*70)

# Calculate feature statistics for each subject
unique_subjects = np.unique(subject_ids)
n_subjects = len(unique_subjects)
n_features = features.shape[1]

# Create subject feature matrices
subject_feature_means = np.zeros((n_subjects, n_features))
subject_feature_stds = np.zeros((n_subjects, n_features))

for i, subject in enumerate(unique_subjects):
    subject_mask = subject_ids == subject
    subject_features = features[subject_mask]
    subject_feature_means[i] = np.mean(subject_features, axis=0)
    subject_feature_stds[i] = np.std(subject_features, axis=0)

# Calculate variability metrics
feature_variability = []

for feat_idx, feat_name in enumerate(feature_names):
    # Between-subject variability (std of subject means)
    between_var = np.std(subject_feature_means[:, feat_idx])
    
    # Within-subject variability (mean of subject stds)
    within_var = np.mean(subject_feature_stds[:, feat_idx])
    
    # Coefficient of variation
    grand_mean = np.mean(subject_feature_means[:, feat_idx])
    
    # Intraclass correlation coefficient approximation
    if within_var > 0:
        icc_approx = between_var**2 / (between_var**2 + within_var**2)
    else:
        icc_approx = 0
    
    feature_variability.append({
        'name': feat_name,
        'between_std': between_var,
        'within_std': within_var,
        'ratio': between_var / (within_var + 1e-10),
        'icc': icc_approx
    })

# Sort by ICC (features that vary more between subjects than within)
feature_variability.sort(key=lambda x: x['icc'], reverse=True)

print("\nFeatures with highest between-subject variability (ICC):")
print("-" * 80)
print(f"{'Feature':<40} {'Between-STD':>12} {'Within-STD':>12} {'ICC':>8}")
print("-" * 80)

for feat in feature_variability[:15]:
    print(f"{feat['name']:<40} {feat['between_std']:>12.3f} {feat['within_std']:>12.3f} {feat['icc']:>8.3f}")

# Visualize the variability pattern
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. ICC distribution
ax = axes[0, 0]
icc_values = [f['icc'] for f in feature_variability]
ax.hist(icc_values, bins=30, edgecolor='black')
ax.set_xlabel('Intraclass Correlation Coefficient')
ax.set_ylabel('Number of Features')
ax.set_title('Distribution of Feature ICC\n(Higher = more between-subject variance)')
ax.axvline(0.5, color='red', linestyle='--', label='ICC=0.5')
ax.legend()

# 2. PCA of all subjects
ax = axes[0, 1]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)

# Color by subject (first 10 subjects for clarity)
colors = plt.cm.tab20(np.linspace(0, 1, 20))
for i, subject in enumerate(unique_subjects[:10]):
    subject_mask = subject_ids == subject
    ax.scatter(features_pca[subject_mask, 0], 
               features_pca[subject_mask, 1], 
               alpha=0.5, s=20, color=colors[i],
               label=subject.split('_')[0][-3:])

ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
ax.set_title('PCA of All Samples (First 10 Subjects)')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

# 3. Subject distance heatmap
ax = axes[1, 0]
from scipy.spatial.distance import pdist, squareform
distances = pdist(subject_feature_means, metric='euclidean')
distance_matrix = squareform(distances)

im = ax.imshow(distance_matrix, cmap='viridis', aspect='auto')
ax.set_title('Pairwise Subject Distances')
# Show only every 4th label for readability
tick_indices = range(0, len(unique_subjects), 4)
ax.set_xticks(tick_indices)
ax.set_yticks(tick_indices)
ax.set_xticklabels([unique_subjects[i].split('_')[0][-3:] for i in tick_indices], rotation=90)
ax.set_yticklabels([unique_subjects[i].split('_')[0][-3:] for i in tick_indices])
plt.colorbar(im, ax=ax, label='Euclidean Distance')

# 4. Feature correlation with PDA across subjects
ax = axes[1, 1]
pda_correlations_by_subject = []

for subject in unique_subjects:
    subject_mask = subject_ids == subject
    subject_features = features[subject_mask]
    subject_pda = targets[subject_mask]
    
    # Calculate correlation for each feature
    correlations = []
    for feat_idx in range(n_features):
        corr = np.corrcoef(subject_features[:, feat_idx], subject_pda)[0, 1]
        correlations.append(corr)
    
    pda_correlations_by_subject.append(correlations)

pda_correlations_by_subject = np.array(pda_correlations_by_subject)

# Plot mean and std of correlations
mean_corrs = np.mean(np.abs(pda_correlations_by_subject), axis=0)
std_corrs = np.std(np.abs(pda_correlations_by_subject), axis=0)

# Sort by mean correlation
sorted_indices = np.argsort(mean_corrs)[-10:]

y_pos = np.arange(len(sorted_indices))
ax.barh(y_pos, mean_corrs[sorted_indices], xerr=std_corrs[sorted_indices], capsize=5)
ax.set_yticks(y_pos)
ax.set_yticklabels([feature_names[i] for i in sorted_indices])
ax.set_xlabel('Mean |Correlation| with PDA')
ax.set_title('Top 10 Features by PDA Correlation\n(averaged across subjects)')

plt.tight_layout()
plt.show()

# Summary statistics
print("\n" + "="*50)
print("SUMMARY STATISTICS")
print("="*50)

# Check if subjects cluster by participant
participants = [s.split('_')[0] for s in unique_subjects]
unique_participants = list(set(participants))

print(f"\nNumber of unique participants: {len(unique_participants)}")
print("Runs per participant:")
for participant in unique_participants:
    runs = [s for s in unique_subjects if s.startswith(participant)]
    print(f"  {participant}: {len(runs)} runs")

# Are the two original test subjects outliers?
original_subjects = ['sub-dmnelf006_run02', 'sub-rtbpd003_run01']
print(f"\nChecking original test subjects:")
for subj in original_subjects:
    if subj in unique_subjects:
        subj_idx = list(unique_subjects).index(subj)
        mean_dist = np.mean(distance_matrix[subj_idx])
        print(f"  {subj}: mean distance to others = {mean_dist:.2f}")
    else:
        print(f"  {subj}: NOT FOUND in processed subjects")
```

```python
# Deep dive into participant differences
print("="*70)
print("PARTICIPANT-SPECIFIC ANALYSIS")
print("="*70)

# Analyze each participant separately
participant_stats = {}

for participant in participants.keys():
    print(f"\n{participant}:")
    
    # Get participant data
    participant_mask = np.array([subj_id.startswith(participant) for subj_id in all_subject_ids])
    p_features = all_features[participant_mask]
    p_targets = all_targets[participant_mask]
    p_ids = all_subject_ids[participant_mask]
    
    # Calculate feature correlations for this participant
    feature_corrs = []
    for i in range(p_features.shape[1]):
        corr = np.corrcoef(p_features[:, i], p_targets)[0, 1]
        feature_corrs.append(abs(corr))
    
    # Find best features for this participant
    best_feat_indices = np.argsort(feature_corrs)[-5:][::-1]
    
    print("  Top 5 features:")
    for idx in best_feat_indices:
        print(f"    {all_feature_names[idx]:<35} |r|={feature_corrs[idx]:.3f}")
    
    # Within-participant model performance
    from sklearn.model_selection import TimeSeriesSplit
    from sklearn.linear_model import Ridge
    from sklearn.preprocessing import StandardScaler
    
    # Use time series split for within-participant
    tscv = TimeSeriesSplit(n_splits=3)
    cv_scores = []
    
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(p_features)
    
    for train_idx, test_idx in tscv.split(X_scaled):
        X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
        y_train, y_test = p_targets[train_idx], p_targets[test_idx]
        
        model = Ridge(alpha=1.0)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        corr = np.corrcoef(y_test, y_pred)[0, 1]
        cv_scores.append(corr)
    
    participant_stats[participant] = {
        'best_feature_corr': max(feature_corrs),
        'mean_cv_score': np.mean(cv_scores),
        'n_samples': len(p_targets)
    }
    
    print(f"  Within-participant CV: {np.mean(cv_scores):.3f} (±{np.std(cv_scores):.3f})")

# Visualize participant differences
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Best feature correlation by participant
ax = axes[0, 0]
participants_list = list(participant_stats.keys())
best_corrs = [participant_stats[p]['best_feature_corr'] for p in participants_list]
participant_names = [p.split('-')[1] for p in participants_list]

bars = ax.bar(participant_names, best_corrs)
ax.set_xlabel('Participant')
ax.set_ylabel('Best Feature |r|')
ax.set_title('Maximum Feature-PDA Correlation by Participant')
ax.set_ylim(0, max(best_corrs) * 1.1)

# Add value labels on bars
for bar, corr in zip(bars, best_corrs):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{corr:.3f}', ha='center', va='bottom')

# 2. Within-participant CV scores
ax = axes[0, 1]
cv_scores = [participant_stats[p]['mean_cv_score'] for p in participants_list]

bars = ax.bar(participant_names, cv_scores)
ax.set_xlabel('Participant')
ax.set_ylabel('Mean CV Correlation')
ax.set_title('Within-Participant Model Performance')
ax.axhline(y=0, color='black', linestyle='-', linewidth=0.5)

# Color bars based on performance
for bar, score in zip(bars, cv_scores):
    if score > 0.3:
        bar.set_color('green')
    elif score > 0.1:
        bar.set_color('yellow')
    else:
        bar.set_color('red')

# 3. Feature correlation heatmap
ax = axes[1, 0]

# Calculate correlation matrix for each participant's top feature
participant_top_features = np.zeros((len(participants_list), len(all_targets)))
feature_names_used = []

for i, participant in enumerate(participants_list):
    participant_mask = np.array([subj_id.startswith(participant) for subj_id in all_subject_ids])
    p_features = all_features[participant_mask]
    p_targets = all_targets[participant_mask]
    
    # Find best feature for this participant
    corrs = [abs(np.corrcoef(p_features[:, j], p_targets)[0, 1]) for j in range(p_features.shape[1])]
    best_idx = np.argmax(corrs)
    
    # Store the feature values
    participant_top_features[i, participant_mask] = all_features[participant_mask, best_idx]
    feature_names_used.append(all_feature_names[best_idx])

# Create correlation matrix between participants' best features
corr_matrix = np.corrcoef(participant_top_features)

im = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
ax.set_xticks(range(len(participant_names)))
ax.set_yticks(range(len(participant_names)))
ax.set_xticklabels(participant_names, rotation=45)
ax.set_yticklabels([f"{p}\n({f.split('_')[0][:10]})" for p, f in zip(participant_names, feature_names_used)])
ax.set_title('Correlation Between Participants\' Best Features')
plt.colorbar(im, ax=ax)

# 4. PDA distribution by participant
ax = axes[1, 1]

for i, participant in enumerate(participants_list):
    participant_mask = np.array([subj_id.startswith(participant) for subj_id in all_subject_ids])
    p_targets = all_targets[participant_mask]
    
    ax.hist(p_targets, bins=30, alpha=0.5, label=participant_names[i], density=True)

ax.set_xlabel('PDA (z-score)')
ax.set_ylabel('Density')
ax.set_title('PDA Distribution by Participant')
ax.legend()

plt.tight_layout()
plt.show()

# Summary recommendations
print("\n" + "="*70)
print("FINAL RECOMMENDATIONS")
print("="*70)

print(f"""
Based on the comprehensive analysis:

1. **Participant-Specific Models are Superior**
   - Within-participant: r = 0.362
   - Between-participant: r = 0.160
   - Individual participants show very different feature-PDA relationships

2. **Key Issues Identified**:
   - Low feature-target correlations (max ~0.15)
   - High between-participant variability
   - Limited sample size per run (144 samples)

3. **Recommended Approach**:
   a) **Short term**: Deploy participant-specific models
      - Train separate model for each participant using their 4 runs
      - Use leave-one-run-out cross-validation
   
   b) **Feature Engineering**: Try these approaches:
      - Sliding window correlations between bands
      - Phase-amplitude coupling
      - Connectivity measures (coherence, PLV)
      - Non-linear features (entropy, fractal dimension)
   
   c) **Data Collection**:
      - Collect more runs per participant
      - Ensure consistent preprocessing across groups
      - Consider real-time PDA measurement validation

4. **Model Architecture**:
   - Use simpler models (Ridge/ElasticNet) due to limited data
   - Strong regularization to prevent overfitting
   - Consider mixed-effects models with participant as random effect

5. **For RTBPD Group**:
   - Process separately with same pipeline
   - Don't mix with DMNELF until you understand group differences
""")

# Save participant-specific results
np.savez('participant_analysis_results.npz',
         participant_stats=participant_stats,
         feature_names=all_feature_names,
         participants=participants_list)

print("\nAnalysis complete. Results saved to 'participant_analysis_results.npz'")
```

```python
# Create deployment-ready models for predictable participants
print("="*70)
print("CREATING DEPLOYMENT-READY MODELS")
print("="*70)

# Focus on the two predictable participants
deployment_participants = ['sub-dmnelf001', 'sub-dmnelf004']

deployment_models = {}

for participant in deployment_participants:
    print(f"\n{participant} Model Development:")
    print("-" * 50)
    
    # Get participant data
    participant_mask = np.array([subj_id.startswith(participant) for subj_id in all_subject_ids])
    p_features = all_features[participant_mask]
    p_targets = all_targets[participant_mask]
    p_ids = all_subject_ids[participant_mask]
    
    # 1. Remove outliers (optional but recommended)
    outlier_mask = np.abs(p_targets) <= 3  # Keep only within 3 std
    p_features_clean = p_features[outlier_mask]
    p_targets_clean = p_targets[outlier_mask]
    p_ids_clean = p_ids[outlier_mask]
    
    print(f"  Removed {np.sum(~outlier_mask)} outliers ({np.sum(~outlier_mask)/len(outlier_mask)*100:.1f}%)")
    
    # 2. Select participant-specific features
    if participant == 'sub-dmnelf001':
        # Use network competition features
        selected_feature_names = ['dmn_index', 'cen_index', 'network_competition',
                                 'beta_alpha_ratio_mean', 'dmn_index_lag1s', 
                                 'cen_index_lag1s', 'network_competition_lag1s',
                                 'dmn_index_lag2s', 'cen_index_lag2s', 'network_competition_lag2s']
    else:  # sub-dmnelf004
        # Use spectral features
        selected_feature_names = ['low_gamma_mean', 'low_gamma_std', 'beta_mean', 
                                 'beta_std', 'beta_alpha_ratio_mean', 'beta_alpha_ratio_std',
                                 'alpha_mean', 'alpha_std', 'theta_mean', 'theta_std']
    
    # Get feature indices
    feature_indices = []
    for feat_name in selected_feature_names:
        if feat_name in all_feature_names:
            idx = list(all_feature_names).index(feat_name)
            feature_indices.append(idx)
    
    X = p_features_clean[:, feature_indices]
    y = p_targets_clean
    
    print(f"  Selected {len(feature_indices)} features")
    
    # 3. Leave-one-run-out cross-validation
    from sklearn.linear_model import Ridge, ElasticNet
    from sklearn.preprocessing import StandardScaler
    from sklearn.metrics import mean_squared_error, r2_score
    
    unique_runs = np.unique(p_ids_clean)
    loro_results = []
    
    for test_run in unique_runs:
        # Split data
        test_mask = p_ids_clean == test_run
        train_mask = ~test_mask
        
        X_train, X_test = X[train_mask], X[test_mask]
        y_train, y_test = y[train_mask], y[test_mask]
        
        # Scale
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        # Train model (Ridge with cross-validated alpha)
        alphas = [0.01, 0.1, 1.0, 10.0, 100.0]
        best_alpha = None
        best_score = -np.inf
        
        for alpha in alphas:
            model = Ridge(alpha=alpha)
            model.fit(X_train_scaled, y_train)
            score = model.score(X_train_scaled, y_train)
            if score > best_score:
                best_score = score
                best_alpha = alpha
        
        # Final model
        model = Ridge(alpha=best_alpha)
        model.fit(X_train_scaled, y_train)
        
        # Predict
        y_pred = model.predict(X_test_scaled)
        
        # Evaluate
        corr = np.corrcoef(y_test, y_pred)[0, 1]
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        
        loro_results.append({
            'run': test_run,
            'correlation': corr,
            'rmse': rmse,
            'alpha': best_alpha
        })
        
        print(f"    {test_run}: r={corr:.3f}, RMSE={rmse:.3f}")
    
    # 4. Train final model on all data
    scaler_final = StandardScaler()
    X_all_scaled = scaler_final.fit_transform(X)
    
    # Use median alpha from CV
    best_alphas = [r['alpha'] for r in loro_results]
    final_alpha = np.median(best_alphas)
    
    final_model = Ridge(alpha=final_alpha)
    final_model.fit(X_all_scaled, y)
    
    # Calculate feature importances (absolute coefficients)
    feature_importances = np.abs(final_model.coef_)
    
    # Store deployment model
    deployment_models[participant] = {
        'model': final_model,
        'scaler': scaler_final,
        'feature_indices': feature_indices,
        'feature_names': [selected_feature_names[i] for i in range(len(feature_indices))],
        'loro_results': loro_results,
        'mean_correlation': np.mean([r['correlation'] for r in loro_results]),
        'feature_importances': feature_importances,
        'outlier_threshold': 3.0
    }
    
    print(f"\n  Final model performance:")
    print(f"    Mean LORO correlation: {deployment_models[participant]['mean_correlation']:.3f}")
    print(f"    Selected alpha: {final_alpha}")

# Visualize deployment model performance
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Deployment Model Analysis', fontsize=14)

for idx, participant in enumerate(deployment_participants):
    model_info = deployment_models[participant]
    
    # Plot 1: LORO performance
    ax = axes[idx, 0]
    runs = [r['run'].split('_')[-1] for r in model_info['loro_results']]
    corrs = [r['correlation'] for r in model_info['loro_results']]
    
    bars = ax.bar(runs, corrs)
    ax.axhline(y=model_info['mean_correlation'], color='red', linestyle='--', 
               label=f'Mean: {model_info["mean_correlation"]:.3f}')
    ax.set_xlabel('Test Run')
    ax.set_ylabel('Correlation')
    ax.set_title(f'{participant} - Leave-One-Run-Out Performance')
    ax.legend()
    
    # Color bars
    for bar, corr in zip(bars, corrs):
        if corr > 0.5:
            bar.set_color('green')
        elif corr > 0.2:
            bar.set_color('yellow')
        else:
            bar.set_color('red')
    
    # Plot 2: Feature importances
    ax = axes[idx, 1]
    importances = model_info['feature_importances']
    feature_names_short = [f.split('_')[0][:8] for f in model_info['feature_names']]
    
    y_pos = np.arange(len(importances))
    ax.barh(y_pos, importances)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(feature_names_short)
    ax.set_xlabel('|Coefficient|')
    ax.set_title(f'{participant} - Feature Importances')

plt.tight_layout()
plt.show()

# Save deployment models
print("\n" + "="*70)
print("SAVING DEPLOYMENT MODELS")
print("="*70)

import pickle

for participant, model_info in deployment_models.items():
    filename = f'deployment_model_{participant}.pkl'
    
    with open(filename, 'wb') as f:
        pickle.dump(model_info, f)
    
    print(f"Saved {participant} model to {filename}")

# Create deployment script template
deployment_code = """
# Deployment code for PDA prediction

import numpy as np
import pickle
from sklearn.preprocessing import StandardScaler

def predict_pda(eeg_features, participant_id, model_file):
    '''
    Predict PDA from EEG features for a specific participant.
    
    Parameters:
    -----------
    eeg_features : dict
        Dictionary containing feature arrays
    participant_id : str
        Participant identifier (e.g., 'sub-dmnelf001')
    model_file : str
        Path to the saved model file
    
    Returns:
    --------
    pda_predictions : array
        Predicted PDA values
    '''
    
    # Load model
    with open(model_file, 'rb') as f:
        model_info = pickle.load(f)
    
    # Extract required features
    feature_vector = []
    for feat_name in model_info['feature_names']:
        if feat_name in eeg_features:
            feature_vector.append(eeg_features[feat_name])
        else:
            raise ValueError(f"Missing required feature: {feat_name}")
    
    # Combine features
    X = np.column_stack(feature_vector)
    
    # Apply outlier check (optional)
    # You may want to check if features are within reasonable ranges
    
    # Scale features
    X_scaled = model_info['scaler'].transform(X)
    
    # Predict
    pda_predictions = model_info['model'].predict(X_scaled)
    
    # Apply outlier threshold to predictions
    pda_predictions = np.clip(pda_predictions, 
                             -model_info['outlier_threshold'], 
                             model_info['outlier_threshold'])
    
    return pda_predictions

# Example usage:
# pda = predict_pda(eeg_features, 'sub-dmnelf001', 'deployment_model_sub-dmnelf001.pkl')
"""

with open('deployment_code.py', 'w') as f:
    f.write(deployment_code)

print("\nCreated deployment_code.py")

# Final summary
print("\n" + "="*70)
print("DEPLOYMENT SUMMARY")
print("="*70)

print(f"""
Successfully created deployment models for:

1. **sub-dmnelf001**:
   - Mean LORO correlation: {deployment_models['sub-dmnelf001']['mean_correlation']:.3f}
   - Uses network competition features
   - Best for participants with strong DMN/CEN patterns

2. **sub-dmnelf004**:
   - Mean LORO correlation: {deployment_models['sub-dmnelf004']['mean_correlation']:.3f}
   - Uses spectral power features
   - Best for participants with clear frequency patterns

Deployment files created:
- deployment_model_sub-dmnelf001.pkl
- deployment_model_sub-dmnelf004.pkl
- deployment_code.py

Next steps:
1. Test models on new data from these participants
2. Monitor prediction quality in real-time
3. Collect more data from similar participants
4. Investigate why other participants are unpredictable
""")
```

```python
def load_pda_flexible(self, pda_file):
    """Load PDA data from CSV file, handling different column naming conventions."""
    pda_df = pd.read_csv(pda_file)
    
    # Check for different column naming conventions
    if 'cen' in pda_df.columns and 'dmn' in pda_df.columns:
        # Format 1: cen, dmn
        pda_signal = pda_df['cen'].values - pda_df['dmn'].values
    elif 'cen_signal' in pda_df.columns and 'dmn_signal' in pda_df.columns:
        # Format 2: cen_signal, dmn_signal
        pda_signal = pda_df['cen_signal'].values - pda_df['dmn_signal'].values
    elif 'pda' in pda_df.columns:
        # Format 3: pre-calculated PDA
        pda_signal = pda_df['pda'].values
    else:
        # Try to find columns containing 'cen' and 'dmn'
        cen_cols = [col for col in pda_df.columns if 'cen' in col.lower() and 'dmn' not in col.lower()]
        dmn_cols = [col for col in pda_df.columns if 'dmn' in col.lower() and 'cen' not in col.lower()]
        
        if cen_cols and dmn_cols:
            print(f"  Found CEN column: {cen_cols[0]}, DMN column: {dmn_cols[0]}")
            pda_signal = pda_df[cen_cols[0]].values - pda_df[dmn_cols[0]].values
        else:
            raise ValueError(f"Could not find CEN and DMN columns. Available columns: {list(pda_df.columns)}")
    
    return pda_signal, pda_df

# Monkey-patch the method into the pipeline
MultiSubjectPDAPipeline.load_pda_flexible = load_pda_flexible

# Now update the process_single_subject method to use this flexible loader
# We need to replace the PDA loading section
def process_single_subject_fixed(self, subject_id, eeg_file, pda_file):
    """Process a single subject's data with flexible PDA loading."""
    print(f"\n{'='*70}")
    print(f"Processing Subject: {subject_id}")
    print(f"{'='*70}")
    
    subject_output_dir = os.path.join(self.output_dir, subject_id)
    os.makedirs(subject_output_dir, exist_ok=True)
    
    # 1. Load and clean EEG data
    print("\n1. Loading and cleaning EEG data...")
    raw_processed, bad_channels, qc_stats = self._clean_eeg_data(
        eeg_file, subject_output_dir
    )
    
    # 2. Load PDA data (UPDATED SECTION)
    print("\n2. Loading PDA data...")
    pda_signal, pda_df = self.load_pda_flexible(pda_file)
    
    # 3. Align PDA with EEG
    print("\n3. Aligning PDA with EEG...")
    pda_aligned_z, pda_time_aligned = self._align_pda_to_eeg(
        pda_signal, raw_processed
    )
    
    # 4. Extract features
    print("\n4. Extracting features...")
    
    # 4a. STFT features
    stft_features = self._extract_stft_features(
        raw_processed, pda_aligned_z, pda_time_aligned
    )
    
    # 4b. LZ complexity features
    lz_features = self._extract_lz_features(
        raw_processed, pda_aligned_z, pda_time_aligned
    )
    
    # 4c. Advanced features
    advanced_features = self._extract_advanced_features(
        raw_processed, stft_features, pda_aligned_z, pda_time_aligned
    )
    
    # 5. Combine all features
    print("\n5. Combining features...")
    all_features = self._combine_features(
        stft_features, lz_features, advanced_features
    )
    
    # 6. Save subject data
    subject_data = {
        'subject_id': subject_id,
        'features': all_features['feature_matrix'],
        'feature_names': all_features['feature_names'],
        'target': pda_aligned_z,
        'time_points': pda_time_aligned,
        'eeg_info': {
            'n_channels': len(raw_processed.ch_names),
            'bad_channels': bad_channels,
            'duration': raw_processed.times[-1]
        }
    }
    
    # Save to file
    np.savez(
        os.path.join(subject_output_dir, f'{subject_id}_processed.npz'),
        **subject_data
    )
    
    print(f"\nSubject {subject_id} processing complete!")
    print(f"Features shape: {subject_data['features'].shape}")
    print(f"Target shape: {subject_data['target'].shape}")
    
    return subject_data

# Apply the fix
MultiSubjectPDAPipeline.process_single_subject = process_single_subject_fixed

# Clear previous results and reprocess
pipeline.subjects_data = {}
pipeline.process_all_subjects(train_runs)
```

```python
# Fix for the alignment issue - just add this function
def combine_subjects_data_fixed(self):
    """Fixed version: Combine data from all subjects for training."""
    print(f"\n{'='*70}")
    print("COMBINING MULTI-SUBJECT DATA")
    print(f"{'='*70}")
    
    all_features = []
    all_targets = []
    all_subject_ids = []
    
    # Debug: Check each subject's data
    print("\nChecking individual subject data:")
    for subject_id, data in self.subjects_data.items():
        print(f"{subject_id}: features {data['features'].shape}, target {data['target'].shape}")
    
    # Get common feature names (use first subject as reference)
    ref_subject = list(self.subjects_data.values())[0]
    common_features = ref_subject['feature_names']
    
    for subject_id, data in self.subjects_data.items():
        # Ensure feature consistency
        if data['feature_names'] == common_features:
            features = data['features']
            target = data['target']
            
            # CRITICAL FIX: Ensure features and targets have same length
            min_length = min(features.shape[0], len(target))
            
            # Trim both to the same length
            features_trimmed = features[:min_length]
            target_trimmed = target[:min_length]
            
            all_features.append(features_trimmed)
            all_targets.append(target_trimmed)
            all_subject_ids.extend([subject_id] * min_length)
            
            print(f"Added {subject_id}: {min_length} samples")
        else:
            print(f"Warning: Feature mismatch for subject {subject_id}")
    
    # Combine arrays
    self.combined_features = np.vstack(all_features)
    self.combined_targets = np.hstack(all_targets)
    self.subject_ids = np.array(all_subject_ids)
    self.feature_names = common_features
    
    print(f"\nCombined data shape:")
    print(f"  Features: {self.combined_features.shape}")
    print(f"  Targets: {self.combined_targets.shape}")
    print(f"  Subjects: {len(np.unique(self.subject_ids))}")

# Apply the fix to the existing pipeline instance
MultiSubjectPDAPipeline.combine_subjects_data = combine_subjects_data_fixed

# Re-run the combination step with the fix
pipeline.combine_subjects_data()
```

```python
# Create a comprehensive performance summary figure
def create_performance_summary(pipeline, test_results):
    """
    Create a summary figure showing model performance.
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('Multi-Subject PDA Prediction Performance Summary', fontsize=16)
    
    # 1. Training vs Test Performance
    ax = axes[0, 0]
    
    # Get training performance
    train_corrs = pipeline.model_results[pipeline.best_model.__class__.__name__]['cv_scores']
    test_corrs = [r['correlation'] for r in test_results if 'correlation' in r]
    
    positions = [1, 2]
    box_data = [train_corrs, test_corrs]
    labels = ['Training\n(Cross-Val)', 'Test\n(Held-Out)']
    
    bp = ax.boxplot(box_data, positions=positions, labels=labels, patch_artist=True)
    bp['boxes'][0].set_facecolor('lightblue')
    bp['boxes'][1].set_facecolor('lightcoral')
    
    ax.set_ylabel('Correlation')
    ax.set_title('Training vs Test Performance')
    ax.grid(True, alpha=0.3)
    
    # Add mean values
    for i, (pos, data) in enumerate(zip(positions, box_data)):
        ax.text(pos, np.mean(data) + 0.02, f'μ={np.mean(data):.3f}', 
                ha='center', va='bottom')
    
    # 2. Per-Subject Test Performance
    ax = axes[0, 1]
    
    if test_results:
        subject_ids = [r['subject_id'] for r in test_results]
        correlations = [r['correlation'] for r in test_results if 'correlation' in r]
        
        bars = ax.bar(range(len(correlations)), correlations)
        ax.set_xlabel('Test Run')
        ax.set_ylabel('Correlation')
        ax.set_title('Test Performance by Run')
        ax.set_xticks(range(len(correlations)))
        ax.set_xticklabels([s.split('_')[-1] for s in subject_ids], rotation=45)
        ax.grid(True, alpha=0.3, axis='y')
        
        # Color code by performance
        for bar, corr in zip(bars, correlations):
            if corr > 0.7:
                bar.set_color('green')
            elif corr > 0.5:
                bar.set_color('orange')
            else:
                bar.set_color('red')
    
    # 3. Feature Type Contribution
    ax = axes[1, 0]
    
    # Categorize features
    feature_categories = {
        'LZ Complexity': 0,
        'Band Powers': 0,
        'Band Ratios': 0,
        'Network Indices': 0,
        'Lagged Features': 0,
        'Other': 0
    }
    
    for feature in pipeline.selected_features:
        if 'lz' in feature.lower():
            feature_categories['LZ Complexity'] += 1
        elif any(band in feature for band in ['theta', 'alpha', 'beta', 'gamma']) and 'ratio' not in feature:
            feature_categories['Band Powers'] += 1
        elif 'ratio' in feature:
            feature_categories['Band Ratios'] += 1
        elif any(net in feature for net in ['network', 'dmn', 'cen']):
            feature_categories['Network Indices'] += 1
        elif 'lag' in feature:
            feature_categories['Lagged Features'] += 1
        else:
            feature_categories['Other'] += 1
    
    # Create pie chart
    categories = [cat for cat, count in feature_categories.items() if count > 0]
    counts = [count for count in feature_categories.values() if count > 0]
    
    wedges, texts, autotexts = ax.pie(counts, labels=categories, autopct='%1.0f%%',
                                      startangle=90)
    ax.set_title('Feature Type Distribution')
    
    # 4. Model Comparison Summary
    ax = axes[1, 1]
    
    model_names = list(pipeline.model_results.keys())
    cv_means = [pipeline.model_results[m]['cv_mean'] for m in model_names]
    full_corrs = [pipeline.model_results[m]['full_correlation'] for m in model_names]
    
    x = np.arange(len(model_names))
    width = 0.35
    
    bars1 = ax.bar(x - width/2, cv_means, width, label='CV Mean', alpha=0.8)
    bars2 = ax.bar(x + width/2, full_corrs, width, label='Full Data', alpha=0.8)
    
    ax.set_xlabel('Model')
    ax.set_ylabel('Correlation')
    ax.set_title('Model Performance Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(model_names, rotation=45)
    ax.legend()
    ax.grid(True, alpha=0.3, axis='y')
    
    # Highlight best model
    best_idx = np.argmax(cv_means)
    bars1[best_idx].set_edgecolor('red')
    bars1[best_idx].set_linewidth(3)
    
    plt.tight_layout()
    plt.savefig(os.path.join(pipeline.output_dir, 'performance_summary.png'),
                dpi=300, bbox_inches='tight')
    plt.show()

# Create the summary
create_performance_summary(pipeline, test_results)
```
